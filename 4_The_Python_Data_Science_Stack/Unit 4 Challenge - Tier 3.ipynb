Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

```python
# Test to find ave price for one borough and year
properties_3[(properties_3['borough']=='Barking & Dagenham') & (properties_3['year']==2018)]['ave_prices'].mean()
```

```python
# Create a function to calculate ave price ratio: ave_price_18 / ave_price_98 
def create_price_ratio_18_95(borough):
    ave_p_2018 = properties_3[(properties_3['borough']==borough) & (properties_3['year']==2018)]['ave_prices'].mean()
    ave_p_1995 = properties_3[(properties_3['borough']==borough) & (properties_3['year']==1998)]['ave_prices'].mean()
    ratio = ave_p_2018/ave_p_1995
    return ratio
```

```python
create_price_ratio_18_95('Barking & Dagenham')
```

```python
### Let's find the ratio of change between 1998 and 2018 for all the boroughs using the function
ratios={}

for i in properties_3['borough'].unique():
    ratios[i]=round(create_price_ratio_18_95(i),3)

ratios
```

```python
# Let's order the ratios dictionary by the boroughs with the highest ave price change from 1998 to 2018
sort_order = sorted(ratios.items(), key=lambda x: x[1], reverse=True)

for i in sort_order:
    print(i[0], i[1])
```

```python
# Let's visualize the ordered boroughs by the highest ave price change
x = [i[0] for i in sort_order]
y = [i[1] for i in sort_order]
plt.xticks(rotation=90)
plt.bar(x,y)
```

```python
# Let's create a borough, year groupby of the df and calculate mean ave price
df_group_mean_1 = properties_3.groupby(['borough', 'year']).mean()
df_group_mean = df_group_mean.reset_index() #resetting index to remove multi indexing
```

```python
# Let's visualize highest ratio borough (Hackney) and lowest ratio borough (Hounslow)
data2 = df_group_mean[df_group_mean['borough'] == 'Hackney']
data2_x = data2['year']
data2_y = data2['ave_prices']
data3 = df_group_mean[df_group_mean['borough'] == 'Hounslow']
data3_x = data3['year']
data3_y = data3['ave_prices']
data4 = df_group_mean[df_group_mean['borough'] == 'City of London']
data4_x = data4['year']
data4_y = data4['ave_prices']
plt.plot(data2_x,data2_y,'g',data3_x,data3_y,'b',data4_x,data4_y,'r')
plt.plot()
plt.legend(['Hackney','Hounslow','City of London'],loc='best')
plt.xlabel('Year')
plt.ylabel('Ave Price')
plt.show()
```

```python
# To get a better idea, let's plot all the borough and the ave prices per year trends
fig,ax = plt.subplots()

for name in df_group_mean['borough'].unique():
    ax.plot(df_group_mean[df_group_mean['borough']==name]['year'],df_group_mean[df_group_mean['borough']==name]['ave_prices'],label=name)

plt.legend(loc='best', bbox_to_anchor=(1.0, 1.0))
plt.show()
```
